```python
import json

with open('xsplot_download.json') as f:
    data = json.load(f)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6.4, 3.2))

li_7 = data[0]
li7_l, =plt.plot(li_7["energy"], li_7["cross section"])

li_6 = data[1]
li6_l, = plt.plot(li_6["energy"], li_6["cross section"])

plt.xscale("log")
plt.yscale("log")

plt.gcf().text(
    0.08, 0.97,
    "Li(n,Xt) cross section",
    weight="bold", fontsize=15,
    fontname="Montserrat",)
subtitle = [
    f"Library: {li_6['Library']}",
    "Made with XSplot.",
]

plt.gcf().text(
    0.08, 0.88, s="\n".join(subtitle), fontsize=6.5,
    fontname="Montserrat",
    url="https://github.com/openmc-data-storage/isotope-xs-plotter"
)

plt.xlabel("Energy (eV)")
plt.ylabel("Cross section (barns)")

plt.annotate("Li6", (1e5, 5e0), color=li6_l.get_color())
plt.annotate("Li7", (8e5, 5e-3),color=li7_l.get_color())
plt.ylim(bottom=min(li_7["cross section"]))

plt.grid(alpha=0.3)
plt.gca().set_axisbelow(True)

plt.gca().spines.right.set_visible(False)
plt.gca().spines.top.set_visible(False)
plt.savefig("cross_section.svg", bbox_inches = "tight")
```
