```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("test_3000nsPW_250PE.csv")

```

```python
plt.hist(df["A_energy [MeV]"], bins = 100, histtype="step", label = "Channel A")
plt.hist(df["B_energy [MeV]"], bins = 100, histtype="step", label = "Channel B")
plt.hist(df["C_energy [MeV]"], bins = 100, histtype="step", label = "Channel C")
print(df.columns)
plt.yscale("log")

plt.ylabel("Counts")
plt.xlabel("Energy [MeV]")
plt.legend()
plt.show()
```

```python

plt.hist(df["Coicidence [MeV]"], bins = 100, histtype="step", label = "Summed Coincidence spectum")

plt.yscale("log")

plt.ylabel("Counts")
plt.xlabel("Energy [MeV]")
# remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.savefig("conincidence_spectrum.svg")
plt.show()
```
